```python
# Create a chat conversation with a system message and a user message

def get_match_score_gpt(job_description, resume):
    conversation = [
        {"role": "system", "content": "You are a Human resources Expert in Software Engineering Recruiting."},
        {"role": "user", "content": "This is a job description that you are hiring for:"},
        {"role": "assistant", "content": job_description},
        {"role": "user", "content": "This is a resume of a candidate who applied for the job:"},
        {"role": "assistant", "content": resume},
        {"role": "user", "content": "Assign a label either 0 or 1. 0 means terrible match, 1 means perfect match. Also provide a short paragrah explaining why you assigned the score."},
    ]

    start = time.perf_counter()
    # Call the model to generate two resumes
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=conversation
    )
    print(f"Time elapsed: {time.perf_counter() - start}")
    return response.choices[0].message["content"]
```

# Iterate

```python
# Input directory containing text files
input_dir = "/Users/andreasloutzidis/Downloads/indeed_data/SWE_junior_le_2years_txt/"

# Output directory for saving processed files
output_dir = "/Users/andreasloutzidis/Downloads/indeed_data/results_senior_jd_w_junior_res/"

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

jd_swe_senior = "/Users/andreasloutzidis/Downloads/indeed_data/jd_SWE_senior/"

with open(jd_swe_senior + "jd_1.txt", "r") as file:
    jd_swe_senior = file.read()
jd_swe_senior = jd_swe_senior.replace("\n", "")
```

```python
i = 0
num = 10
while i < num:
    # Iterate through each text file in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith(".txt"):
            input_file_path = os.path.join(input_dir, filename)
            output_file_path = os.path.join(output_dir, filename)

            # Read the text file
            with open(input_file_path, "r", encoding="utf-8") as file:
                file_contents = file.read()
#             print(f"file contents: {file_contents[:100]}")
            # Process the file contents using your custom function
            try:
                processed_data = get_match_score_gpt(job_description=jd_swe_senior, resume=file_contents)
                # Save the processed data to the output directory with the same filename
                with open(output_file_path, "w", encoding="utf-8") as output_file:
                    output_file.write(processed_data)
                print(f"Processed {filename} and saved to {output_file_path}")
            except Exception as e:
                print(f"ERROR processing {filename}")
                continue
            print(f"Progress = {i}/{num}")
    i += 1
        
print("Processing completed.")
```
