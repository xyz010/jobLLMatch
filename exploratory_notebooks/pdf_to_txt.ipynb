```python
import pdfplumber
import re
import os

# Define a function to extract the work experience section from a PDF resume
def extract_work_experience_with_stop(pdf_file_path, stop_keywords):
    work_experience = ""
    with pdfplumber.open(pdf_file_path) as pdf:
        text = ""
        start_flag = False

        for page in pdf.pages:
            page_text = page.extract_text()
            text += page_text

            if not start_flag:
                # Check if the work experience section has started
                start_match = re.search(r"(Work Experience|Experience|Professional Experience)", page_text, re.IGNORECASE)
                if start_match:
                    start_flag = True

            if start_flag:
                # Check if any of the stop keywords are found
                for keyword in stop_keywords:
                    if re.search(keyword, page_text, re.IGNORECASE):
                        return work_experience

                # Append the current page's text to the work experience
                work_experience += page_text

    return work_experience

# Provide the path to the PDF file
pdf_file_path = "resume.pdf"


```

```python
resume_dir = "/Users/andreasloutzidis/Downloads/indeed_data/swe_big_resume_bucket_pdf"
resume = "A--Shaw.pdf"
# Provide the path to the PDF file
pdf_file_path = os.path.join(resume_dir, resume)


# Define the stop keywords
stop_keywords = ["Skills", "Education", "Projects"]

# Extract and print the work experience section stopping at specified keywords
work_experience_text = extract_work_experience_with_stop(pdf_file_path, stop_keywords)
print(work_experience_text)
```
